```python
import os

def find_xml_files(directory):
    xml_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.xml'):
                xml_files.append(os.path.join(root, file))
    return xml_files
```

```python
xml_list = []
for i in range(5):
    print(i)
    xml_list.append([])
    directory_path = str('C:/Users/User/Matlab_data/인도보행 영상/Surface_Masking/Surface_' + str(i + 1))
    print(directory_path)
    xml_files = find_xml_files(directory_path)
    for j in range(len(xml_files)):
        xml_list[i].append(xml_files[j][xml_files[j].find("\\", xml_files[j].find('\\') + 1) + 1:])
```

```python
import xmltodict
import json
```

```python
cnt = 0
for i in range(5):

    for j in range(len(xml_list[i])):
        cnt += 1
        cnt2str = "{:03}".format(cnt)
        path = f"Surface_Masking/Surface_{i+1}/Surface_{cnt2str}/{xml_list[i][j]}"
        print(path)

        try:
            with open(path) as xml_file:
                data_dict = xmltodict.parse(xml_file.read())
            json_data = json.dumps(data_dict)
            json_path = f"Surface_Masking/Surface_{i+1}/Surface_{cnt2str}/{xml_list[i][j][:-3]}json"
            with open(json_path, "w") as json_file:
                json_file.write(json_data)
        except FileNotFoundError:
            print(f"File not found: {path}")
        
        except xmltodict.expat.ExpatError as e:
            print(f"XML parsing error in file {path}: {e}")
        
        except Exception as e:
            print(f"An unexpected error occurred with file {path}: {e}")
```
